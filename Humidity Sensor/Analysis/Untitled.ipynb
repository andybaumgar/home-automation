```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.rcParams['figure.dpi']= 227

# df = pd.read_csv("DATALOG_2.TXT")
df = pd.read_csv("DATALOG_20190919.TXT")
df = df.iloc[::60]


print (df.shape)

size = len(df.index)
day = 60*24
df = df.iloc[size-4*day:size] 

print (df.shape)

df['time'] = pd.to_datetime(df.time, unit="s")

fig, ax = plt.subplots(figsize=(15,7))

df.plot(x="time", y=["temperature", "humidity"], ax=ax, linewidth=.5)

ax.xaxis.set_major_locator(mdates.HourLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H'))
ax.xaxis.set_label("Time(hours)")
plt.show()
```
